```python
import matplotlib.pyplot as plt
import numpy as np
import time
import copy
from scipy import signal
from scipy import interpolate
import os
from scipy.signal import find_peaks
from scipy.signal import savgol_filter

%matplotlib notebook
```

```python
# 归一化 
def osc_pre(y,key):
    yy = []
    yy = copy.copy(y)
    
    if key == 0: # 正
        yy = (yy )
        y0mean=np.mean(yy[0:100])
        yemean=np.mean(yy[-100:-1])
    elif key == 1: #负
        yy = (yy) *-1
        yemean=np.mean(yy[0:100])
        y0mean=np.mean(yy[-100:-1])
    
    
    if yemean < y0mean:
        yy = (yy - y0mean)*-1
    else:
        yy = (yy - y0mean)
        
        
    if key == 0: # 正
        yy = (yy )
        y0mean=np.mean(yy[0:100])
        yemean=np.mean(yy[-100:-1])
    elif key == 1: #负
        yy = (yy) *-1
        yemean=np.mean(yy[0:100])
        y0mean=np.mean(yy[-100:-1])
    
    yy = (yy) /yemean
#     yy = [0 if i < 0 else i for i in yy]
      
    return yy
```

```python
# 归一化 
def osc_prex(y,key):
    yy = []
    yy = copy.copy(y)
    yyy = []
    yyy = copy.copy(y)

    y0mean=np.mean(yy[-100:-10])

    y_min = np.abs(np.min(yy)-y0mean)
    y_max = np.abs(np.max(yy)-y0mean)
    y_min = np.min([y_min,y_max])
    yyy = (np.abs(yyy-y0mean)).tolist()
    index = yyy.index(y_min)  

    yemean=np.mean(yy[index-200:index-100])


    if yemean < y0mean:
        yy = (yy - y0mean)*-1
    else:
        yy = (yy - y0mean)

    yyy = copy.copy(yy)
    y0mean=np.mean(yy[-100:-10])

    y_min = np.abs(np.min(yy)-y0mean)
    y_max = np.abs(np.max(yy)-y0mean)
    y_min = np.min([y_min,y_max])
    yyy = (np.abs(yyy-y0mean)).tolist()
    index = yyy.index(y_min)  

    yemean=np.mean(yy[index-200:index-100])  

    yy = (yy) /yemean
      
    return yy
```

```python
# 通过diff find 上升沿与下降沿
def diff_find(y):
    y_diff = np.diff(y)
    t = range(len(y_diff))
    return t,y_diff
```

```python
# find square
def square_find(y):
    dt1,dy1 = diff_find(y)
#     dn = (dy1.tolist().index(max(dy1)))
    
    peaks, _ = find_peaks(dy1, height=0.1)
    for i in peaks:
        if y[i+1000] >0.6 and i+int(width/k)<int(len(dy1)):
            dn = i
    print(dn)
    
    dn2 = int(dn + width/k)
    dtx = [i/(1/k) for i in range(dn2 - dn)]
    dyx = y[dn:dn2]
    
    return dtx,dyx
```

```python
# find falledge
def falledge_find(y):
    dt1,dy1 = diff_find(-y)
    peaks, _ = find_peaks(dy1, height=0.1)

    for i in peaks:
#         if y[i+1000] <0.1 and i+int((period-width)/k)<int(len(y)):
        if y[i+1000] <0.1:
            dnn = i
    print(dnn)
    
#     dn2 = int(dnn + (period-width)/k)
    dtx = [i/(1/k) for i in range(len(y)-dnn)]
    dyx = y[dnn:len(y)]
   
    return dtx,dyx
```

## up vs fall

```python
def overturn(yy):
    y = copy.copy(yy)
    y = 1-y
    return y
```

```python
def diff_compare(yy1,yy2):
    y1 = copy.copy(yy1)
    y2 = copy.copy(yy2)
    n = np.min([len(y1),len(y2)])
    t = [i/(1/k) for i in range(n)]
    
    return t, y1[0:n]-y2[0:n]
```

```python
# data process
# up
tx1,ty1 = square_find(osc_prex(amplist1,0))
tx2,ty2 = square_find(osc_prex(amplist2,0))
# tx3,ty3 = square_find(osc_pre(amplist3,0))
# tx4,ty4 = square_find(osc_pre(amplist4,0))

# fall
tfx1,tfy1 = falledge_find(osc_prex(amplist1,0))
tfx2,tfy2 = falledge_find(osc_prex(amplist2,0))
# tfx3,tfy3 = falledge_find(osc_pre(amplist3,0))
# tfx4,tfy4 = falledge_find(osc_pre(amplist4,0))

tfy1 = overturn(tfy1)
tfy2 = overturn(tfy2)
# tfy3 = overturn(tfy3)
# tfy4 = overturn(tfy4)

# vs

t1,y1 = diff_compare(ty1,tfy1)
t2,y2 = diff_compare(ty2,tfy2)
# t3,y3 = diff_compare(ty3,tfy3)
# t4,y4 = diff_compare(ty4,tfy4)
```
