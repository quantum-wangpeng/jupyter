```python
Nstates = 3
#升降算符
a = destroy(Nstates)
a_dagger = create(Nstates)  
I = qeye(Nstates)
I2= tensor(I,I)
a1 = tensor(a,I)
a1d = a1.dag()  
a2 = tensor(I,a)
a2d = a2.dag()
#单比特初始裸态
ground = basis(Nstates,0)
excited = basis(Nstates,1)
leakage = basis(Nstates,2)
sx1 = a1 + a1d
sy1 = 1j*(a1d - a1)
sz1 = I2 - 2*a1d*a1
sx2 = a2 + a2d
sy2 = 1j*(a2d - a2)
sz2 = I2 - 2*a2d*a2
```

## 系统哈密顿量

```python
def Hamiltonian(device_paras):
    w1=device_paras['w'][0]*2.0*np.pi
    w2=device_paras['w'][1]*2.0*np.pi
    alpha1=device_paras['alpha'][0]*10**(-3)*2.0*np.pi
    alpha2=device_paras['alpha'][1]*10**(-3)*2.0*np.pi

    g=device_paras['g']*10**(-3)*2.0*np.pi

    H0 = w1*a1d*a1+w2*a2d*a2 +\
          alpha1/2.0* a1d * a1d * a1 * a1+alpha2/2.0* a2d * a2d * a2 * a2
    HI = -g*(a1-a1d)*(a2 - a2d)
    H = H0+HI
    return H0 ,H 
    
```

## 本征态

```python
dim = [3, 3]
def energy_info_index(i):
    if i ==-1:
        i = sum(dim)-len(dim)
    energy_info_q = {'dim':dim, 'exci_num': i, 'bas_list':[], 'bas':[], 'bas_name_list':[], 'bas_index':[]}
    for bas in state_number_enumerate(energy_info_q['dim'], excitations=energy_info_q['exci_num']):
        energy_info_q['bas_list'].append(state_number_qobj(energy_info_q['dim'], bas))
        energy_info_q['bas_name_list'].append(''.join(map(str,bas)))
        energy_info_q['bas'].append(bas)
        if i == 0:
            energy_info_q['bas_index'].append(0)
            break
        else:
            energy_info_q['bas_index'].append(state_number_index(energy_info_q['dim'], bas))
    energy_info_q['bas_num'] = len(energy_info_q['bas_list'])
    return energy_info_q
```

```python
def eigensolve(paras_dict):
    
    H0,H = Hamiltonian(paras_dict) # 获得裸态和缀饰态的哈密顿量
    # 使用eigenstates()函数计算哈密顿量的本征态和本征矢
    ei_states = H.eigenstates() # 得到dressed states的能级变化
    ei_energy = ei_states[0]
    ei_vector = ei_states[1]
    
    ei_states0 = H0.eigenstates() # 得到bare states的能级变化
    ei_energy0 = ei_states0[0]
    ei_vector0 = ei_states0[1]
    
    states_order = ei_vector.copy()
    states0_order = ei_vector.copy()
    energy_order = ei_energy.copy()
    energy0_order = ei_energy.copy()
    
    
    bas_list = energy_info_index(np.prod(dim))['bas_list']
    for n,vector in enumerate(ei_vector0):
        try:
            index = bas_list.index(vector)
            states_order[index] = ei_vector[n]
            states0_order[index] = ei_vector0[n]
            energy_order[index] = ei_energy[n]
            energy0_order[index] = ei_energy0[n]            
            
        except:
            pass
    
    return states_order, states0_order , energy_order, energy0_order
```

```python

def energy_level_diagiam(paras_dict):
    parameter = []
    for w in w_list:
        paras_dict['w'][1] = w
        parameter.append(copy.deepcopy(paras_dict))
    energy_result = parallel_map(eigensolve, parameter, progress_bar=True,num_cpus=24)   
    energy_list = np.array([energy[2] for energy in energy_result])
    energy0_list = np.array([energy[3] for energy in energy_result]) 
    
    return energy_list, energy0_list
    
   
```

## 驱动脉冲

```python
def drive_pulseX1(t, args):  
    t_g = args['gate time']
    amp = args['q1 amp']
    w_d = args['q1 drive frequency']*2*np.pi
    alpha = device_paras['alpha'][0]*2*np.pi*10**(-3)
    detune = args['q1 detune']*2*np.pi*10**(-3)
    lambda1 = args['drag weight']
    phi = args['q1 phi']*2*np.pi*10**(-3)
    I=np.cos(phi)
    Q=np.sin(phi)
    X0=amp * (1-np.cos(2*np.pi*t/tg))/2 
    Y0=- amp * np.pi/tg * np.sin(2*np.pi*t/tg)/alpha*lambda1
    X = (X0* I +Y0* Q)*np.cos((w_d+detune) * t) 
    return X

def drive_pulseY1(t, args):  
    t_g = args['gate time']
    amp = args['q1 amp']
    w_d = args['q1 drive frequency']*2*np.pi
    alpha = device_paras['alpha'][0]*2*np.pi*10**(-3)
    detune = args['q1 detune']*2*np.pi*10**(-3)
    lambda1 = args['drag weight']
    phi = args['q1 phi']*2*np.pi*10**(-3)
    I=np.cos(phi)
    Q=np.sin(phi)
    X0=amp * (1-np.cos(2*np.pi*t/tg))/2 
    Y0=- amp * np.pi/tg * np.sin(2*np.pi*t/tg)/alpha*lambda1
    Y = (Y0* I -X0* Q)*np.cos((w_d+detune) * t) 
    return Y
```

## 单比特QPT

```python
rho0 = fock_dm(3,0)
rho1 = fock_dm(3,1)
rho2 = ket2dm((basis(3,0)+basis(3,1))/np.sqrt(2))
rho3 = ket2dm((basis(3,0)+1j*basis(3,1))/np.sqrt(2))

rho_basis = [rho0, rho1, rho2, rho3]
rho_in = [tensor(rho,rho0) for rho in rho_basis]

op_basis = [[qeye(2), sigmax(), sigmay(), sigmaz()]]
U_rho = spre(sigmax()) * spost(sigmax().dag())
chi_ideal = qpt(U_rho, op_basis)


def qpt1(U):

    
    rho_out=[(U * rho_in[i]*U.dag()).ptrace(0)[[0,1],:][:,[0,1]] for i in range(len(rho_in))]
    rho_basis_mat = np.hstack([mat2vec(rho.full()) for rho in rho_basis])
    rho_out_mat = np.hstack([mat2vec(rho) for rho in rho_out])
    Gamma = linalg.solve(rho_basis_mat.T, rho_out_mat.T)
    Gamma = Gamma.T
    vecGamma = mat2vec(Gamma)
    
    op_basis = [qeye(2), sigmax(), sigmay(), sigmaz()]
    EE_ops = [sprepost(E1, E2.dag()) for E1 in op_basis for E2 in op_basis]
    Beta = np.hstack([mat2vec(EE.full()) for EE in EE_ops])
    vecChi = linalg.solve(Beta, vecGamma)
    Chi = vec2mat(vecChi)
    return Chi
```

## 保真度计算

```python
def Fidelity_X(U):
    
    f = lambda phi :  1-abs(np.trace(np.dot((rz(phi)*U).dag(),(rz(phi)*U)))+(abs(np.trace(np.dot(sigmax().dag(),(rz(phi)*U))))** 2))/6
    res = minimize_scalar(f, method='brent')
    return res.fun
```

```python
def par_X1(device_paras):

    
    global args1
    args1 = {}  
    args1['gate time'] = tg
    args1['q1 drive frequency'] = device_paras['w'][0]
    args1['drag weight']=0.5
    
    H0,H = Hamiltonian(device_paras)
    states, states0, energy, energy0 =eigensolve(H0,H)

    

    
    
    #校准Q1，校准时Q2置0
    detune_0=(energy[state_number_index(energy_info['dim'], [1, 0])]-\
            energy[state_number_index(energy_info['dim'], [0, 0])]-\
            energy0[state_number_index(energy_info['dim'], [1,0])]+\
            energy0[state_number_index(energy_info['dim'], [0, 0])])/(2*np.pi*10**(-3))
    x0 = [detune_0,0.21]
    bounds = ((x0[0]-10,x0[0]+10),(x0[1]-0.07,x0[1]+0.07) )
    start = time.time();count=0;error_list = [];
    result1 = minimize(Evolution_q1, x0 ,args=(H,states,0,),bounds=bounds, method='SLSQP',options={'ftol': 1e-06})
    
    #Q2置1,演化Q1
    
    error = Evolution_q1(result1.x,H,states,1)

    

    return result1.fun, error
```

```python
def pulse_2(t_list, paras_dict):
    # 计算上升沿下降沿对应的时间戳
    pulse_len = paras_dict['pulse_len']
    sigma = paras_dict['sigma']
    x_idle = device_paras['w'][1]
    x_work = paras_dict['w2_work']
    flattop_start = 3*sigma
    flattop_end = pulse_len - 3*sigma


    y_list = (x_work-x_idle) * 1/2*( erf( (t_list-flattop_start)/(np.sqrt(2)*sigma) ) - \
                                erf( (t_list-flattop_end)/(np.sqrt(2)*sigma) ) ) + x_idle
    return y_list
```

```python
def pulse_2_xy(t_list, paras_dict):
    # 计算上升沿下降沿对应的时间戳

    sigma = paras_dict['sigma']
    pulse_len = paras_dict['pulse_len']
    amp = paras_dict['amp']
    w_d = paras_dict['w'][1]*2*np.pi
    detune = paras_dict['detune']*2*np.pi*10**(-3)

    flattop_start = 3*sigma
    flattop_end = pulse_len - 3*sigma


    y_list = amp * 1/2*( erf( (t_list-flattop_start)/(np.sqrt(2)*sigma) ) - \
                                erf( (t_list-flattop_end)/(np.sqrt(2)*sigma) ) ) *np.cos((w_d+detune) * t_list)
    

    
    
    
    return y_list
```

## 含时哈密顿量

```python
def Hamiltonian_t(t_list, paras_dict):
    w1=device_paras['w'][0]*2.0*np.pi
    w2=pulse_2(t_list, paras_dict)*2.0*np.pi
    alpha1=device_paras['alpha'][0]*10**(-3)*2.0*np.pi
    alpha2=device_paras['alpha'][1]*10**(-3)*2.0*np.pi

    g=device_paras['g']*10**(-3)*2.0*np.pi

    H1 = w1*a1d*a1
    H2 = [a2d*a2, w2]

    
    H_alpha = alpha1/2.0* a1d * a1d * a1 * a1+alpha2/2.0* a2d * a2d * a2 * a2
    HI = -g*(a1-a1d)*(a2 - a2d)
    H = [H1, H2, H_alpha, HI]

    


    

    

    return H
    
    
```

```python
def Hamiltonian_t_trans(t_list, paras_dict,):
    states, states0, energy, energy0 = eigensolve(paras_dict)
    
    
    w1=device_paras['w'][0]*2.0*np.pi
    w2=pulse_2(t_list, paras_dict)*2.0*np.pi
    alpha1=device_paras['alpha'][0]*10**(-3)*2.0*np.pi
    alpha2=device_paras['alpha'][1]*10**(-3)*2.0*np.pi

    g=device_paras['g']*10**(-3)*2.0*np.pi

    H1 = w1*a1d*a1
    H_alpha = alpha1/2.0* a1d * a1d * a1 * a1+alpha2/2.0* a2d * a2d * a2 * a2
    HI = -g*(a1-a1d)*(a2 - a2d)
    
    
    H_static = (H1+ H_alpha+ HI).transform(states)
    
    H2 = [(a2d*a2).transform(states), w2]
    
    H = [H_static, H2]

    


    

    

    return H
    
    
```

## 含时演化

```python
def states_t(paras_dict,state_start_num,cphase_cal = False):
    n = sum(state_start_num)
    state_index_list_r = energy_info_index(n)['bas_index']+energy_info_index(n-1)['bas_index']
    state_index_list = [item for item in state_index_list_r if state_index_list_r.count(item) == 1]
#     state_index_list = energy_info_1['bas_index']
    t_list = pulse_paras['t_list']
    states, states0, energy, energy0 = eigensolve(paras_dict)
    Ht = Hamiltonian_t(t_list, paras_dict)
    # 计算末态布局数
    state_start = states[state_number_index(dim, state_start_num)]
    states_set = [ket2dm(states[index]) for index in state_index_list]
    psi_final = mesolve(Ht, state_start, t_list).states
    P_list = [expect(state,psi_final) for state in states_set]
    
    state_name_list = [energy_info_index(np.prod(dim))['bas_name_list'][i] for i in state_index_list]
    
    
    
    
    # 计算受控相位
    if cphase_cal == True:
        psi_basis = [[states[state_number_index(dim, [i,j])]  for j in range(3)] for i in range(3)]
        
        b1x = 1/np.sqrt(2) * (psi_basis[1][0] + psi_basis[1][1])
        b0x = 1/np.sqrt(2) * (psi_basis[0][0] + psi_basis[0][1])
        psi1x_list = mesolve(Ht, b1x, t_list).states
        psi0x_list = mesolve(Ht, b0x, t_list).states
        
        
        phase1x_arr = np.array([cmath.phase(measure(states,psi1x,[1] ,[(1,0)] )) for psi1x in psi1x_list])
        phase0x_arr = np.array([cmath.phase(measure(states,psi0x,[1] ,[(1,0)] )) for psi0x in psi0x_list])
        cphase = phase1x_arr - phase0x_arr
        cphase = np.arctan2(np.sin(cphase), np.cos(cphase))
        return P_list,cphase,state_name_list
    else:
        cphase = np.zeros(len(t_list))
        return P_list,cphase,state_name_list
    
    
    
    
```

```python
def states_t_trans(paras_dict,state_start_num,cphase_cal = False):
    n = sum(state_start_num)
    state_index_list_r = energy_info_index(n)['bas_index']+energy_info_index(n-1)['bas_index']
    state_index_list = [item for item in state_index_list_r if state_index_list_r.count(item) == 1]
    t_list = pulse_paras['t_list']

    Ht = Hamiltonian_t_trans(t_list, paras_dict)
   
    # 计算末态布局数
#     state_start = state_number_qobj(dim, state_start_num)
    state_start = basis(dim,[1]+[1])
    states_set = [ket2dm(energy_info_index(-1)['bas_list'][index]) for index in state_index_list]
    psi_final = mesolve(Ht, state_start, t_list).states
    P_list = [expect(state,psi_final) for state in states_set]
    
    state_name_list = [energy_info_index(-1)['bas_name_list'][i] for i in state_index_list]
    
    
    
    
    # 计算受控相位
    if cphase_cal == True:
        b100x = 1/np.sqrt(2) * (basis(dim,[1]+[0])+basis(dim,[1]+[1]))
        b000x = 1/np.sqrt(2) * (basis(dim,[0]+[0])+basis(dim,[0]+[1]))
        psi100x_list = mesolve(Ht, b100x, t_list).states
        psi000x_list = mesolve(Ht, b000x, t_list).states
        phase100x_arr = np.array([cmath.phase(psi100x.ptrace(len(dim)-1)[1,0]) for psi100x in psi100x_list])
        phase000x_arr = np.array([cmath.phase(psi000x.ptrace(len(dim)-1)[1,0]) for psi000x in psi000x_list])
        cphase = phase100x_arr - phase000x_arr
        cphase = np.arctan2(np.sin(cphase), np.cos(cphase))
        return P_list,cphase,state_name_list
    else:
        cphase = np.zeros(len(t_list))
        return P_list,cphase,state_name_list
    
    
    
    
```

## 泄漏计算

```python
def par_leak(paras_dict):
    
    t_list = np.linspace(0,paras_dict['pulse_len'],300)
    
    H0,H = Hamiltonian(paras_dict)
    states, states0, energy, energy0 =eigensolve(H0,H)
    
    Ht = Hamiltonian_t(t_list, paras_dict)
    
    b11 = states[state_number_index(energy_info['dim'], [1, 1])]
    psi = mesolve(Ht, b11, t_list,args = paras_dict).states[-1]
    amp11 = np.abs(psi.overlap(b11))**2

    return amp11
```

```python
detune0 = 200
pulse_paras={'pulse_len':30,'sigma':0.3,'detune':detune0}

device_paras['g'] = np.pi/(np.sqrt(2)*copy.deepcopy(pulse_paras['pulse_len']))/(2*np.pi*10**(-3))*3

parameter=[]

amp0 = np.sqrt(2*detune0*10**(-3)*(device_paras['w'][1]-device_paras['w'][0]+device_paras['alpha'][0]*10**(-3)))*2*np.pi
# amp_list=np.linspace(-1.5,0.1,30)+amp0
amp_list=np.linspace(0.0,1,30)

# pulselen_list=np.linspace(30,90,30)+pulse_paras['pulse_len']
pulselen_list=np.linspace(10,150,30)

for pulse_len in pulselen_list:
    for amp in amp_list:
        pulse_paras['pulse_len']=pulse_len
        pulse_paras['amp']=amp
        CZ_paras = {**device_paras,**pulse_paras}
        parameter.append(copy.deepcopy(CZ_paras))
        
result=parallel_map(par_leak, parameter, progress_bar=True,num_cpus=18)
```

## 受控相位计算

```python
def par_cphase(paras_dict):
    t_list = np.linspace(0,paras_dict['pulse_len'],300)
    
    Ht = Hamiltonian_t(t_list, paras_dict)
    dim=energy_info['dim']

    b1x = 1/np.sqrt(2) * (basis(dim,[1,0])+basis(dim,[1,1]))
    b0x = 1/np.sqrt(2) * (basis(dim,[0,0])+basis(dim,[0,1]))
    psi1x = mesolve(Ht, b1x, t_list,args = paras_dict).states[-1]
    psi0x = mesolve(Ht, b0x, t_list,args = paras_dict).states[-1]
    phase1x = cmath.phase(psi1x.ptrace(1)[1,0])
    phase0x = cmath.phase(psi0x.ptrace(1)[1,0])
    
    cphase = phase1x - phase0x
    cphase = np.arctan2(np.sin(cphase), np.cos(cphase))
    
    return cphase
```

## 保真度计算

```python
def Fidelity_CZ(U):
    U_CZ = CPhase(theta=np.pi)
#     error =1-abs(np.trace(np.dot(U.dag(),U))+(abs(np.trace(np.dot(U_CZ.dag(),U)))** 2))/20
    I = qeye(2)
    
    f = lambda phi :  1-abs(np.trace(np.dot((tensor(rz(phi[0]),I)*tensor(I,rz(phi[1]))*U).dag(),(tensor(rz(phi[0]),I)*tensor(I,rz(phi[1]))*U)))+(abs(np.trace(np.dot(U_CZ.dag(),(tensor(rz(phi[0]),I)*tensor(I,rz(phi[1]))*U))))** 2))/20

    res = minimize(f,[0,0],method='SLSQP')
    error =res.fun
    return error
```

```python
def par_fidelity(paras_dict):
    dim=energy_info['dim']
    t_list = np.linspace(0,paras_dict['pulse_len'],300)
    H0,H = Hamiltonian(paras_dict)
    states, states0, energy, energy0 =eigensolve(H0,H)

    Ht = Hamiltonian_t(t_list, paras_dict)
    U_uncali = propagator(Ht, t_list,args = paras_dict)[-1]
    
    
    idx00 = state_number_index(dim, [0,0]); 
    idx01 = state_number_index(dim, [0,1]); 
    idx10 = state_number_index(dim, [1,0]); 
    idx11 = state_number_index(dim, [1,1]);
#     U = Qobj(U_cali[[0,idx01,idx10,idx11],:][:,[0,idx01,idx10,idx11]], dims=[[2,2],[2,2]])
    basis_calcu = [idx00,idx01,idx10,idx11]
    U=np.zeros([4,4],dtype='complex128')
    for i in range(4):
        for j in range(4):
            U[i][j] = (states[basis_calcu[i]].dag()*U_uncali*states[basis_calcu[j]]).full()[0][0]
    U=Qobj(U, dims=[[2,2],[2,2]])
    error = Fidelity_CZ(U)
  
    
    
    
    return error
```

```python
error=parallel_map(par_fidelity, parameter, progress_bar=True,num_cpus=24)
```

```python
def calibration_F(paras,device_paras,n):
    pulse_paras = {'sigma':1.2}
    pulse_paras['pulse_len'] = paras[0]
    pulse_paras['w2_work'] = paras[1]
    CZ_paras = {**device_paras,**pulse_paras}

    
    error = par_fidelity(CZ_paras,n)
    error_list.append(error)
    return error 
    
    
    
```

```python
def calibration_leak(paras,device_paras):
    pulse_paras = {'sigma':1.2}
    pulse_paras['pulse_len'] = paras[0]
    pulse_paras['w2_work'] = paras[1]
    CZ_paras = {**device_paras,**pulse_paras}
    
    
    error = 1-par_leak(CZ_paras)+(np.pi-abs(par_cphase(CZ_paras)))/np.pi
    
    error_list.append(error)
    return error
    
    
    
```

```python
def calibration_cphase(paras,device_paras):
    pulse_paras = {'sigma':1.2}
    pulse_paras['pulse_len'] = paras[0]
    pulse_paras['w2_work'] = paras[1]
    CZ_paras = {**device_paras,**pulse_paras}

    
    error = np.pi-abs(par_cphase(CZ_paras))
    
    error_list.append(error)
    return error
    
    
    
```

```python
def _EPGback(argi):
    global count

    if count%1 == 0:
        end=time.time()
        print('iter:{:d}  error:{:f}  {}  {:2f}'.format(count, error_list[-1] ,argi,end-start)) 
    count += 1
```
